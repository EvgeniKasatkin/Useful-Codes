```python
import math as m
import numpy as np
import pandas as pd

test = pd.read_csv('/1_calculate_percentile_recall/test.tsv')
test = test.sort_values(by = 'predict', ascending = False)
test = test.reset_index(drop = True)
test['id'] = test.index
test['quantile'] = test['id'].apply(lambda x: m.floor((x)/(test['id'].quantile(.1)+1)))
stat = pd.DataFrame({
    'total': test.groupby(['quantile'])['quantile'].count(), 
    'sum_of_target': test.groupby(['quantile'])['target'].agg(np.sum),
    'min_of_score': test.groupby(['quantile'])['predict'].agg(np.min),
    'max_of_score': test.groupby(['quantile'])['predict'].agg(np.max),
})
stat['sum_target_percentile'] =  stat['sum_of_target'].cumsum()
stat['recall'] = round(100*stat['sum_target_percentile']/stat['sum_of_target'].sum(),2)
stat
```
